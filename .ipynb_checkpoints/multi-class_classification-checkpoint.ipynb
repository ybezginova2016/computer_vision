# Multi-class Classification

Вычисление функции потерь не поменяется. Только называться она будет CE (англ. cross-entropy, «кросс-энтропия»):



Функция активации, которая здесь подойдёт, называется SoftMax (англ. «мягкий максимум»). Она принимает несколько выходов сети и возвращает вероятности, сумма которых равна единице.

```
p1 = SoftMax(z1) = e^z1 / (e^z1 + e^z2 + e^z3)
p2 = SoftMax(z2) = e^z2 / (e^z1 + e^z2 + e^z3)
p3 = SoftMax(z3) = e^z3 / (e^z1 + e^z2 + e^z3)

```

### SoftMax - функция активации на последнем слое в задаче многоклассовой классификации:

```
p1 + p2 + p3 = SoftMax(z1) + SoftMax(z2) + SoftMax(z3) =

= e^z1 / (e^z1 + e^z2 + e^z3) + e^z2 / (e^z1 + e^z2 + e^z3) + e^z3 / (e^z1 + e^z2 + e^z3) =

= (e^z1 + e^z2 + e^z3) / (e^z1 + e^z2 + e^z3) = 1
```



- Сумма всех выходов SoftMax равна единице. На выходе получаются вероятности классов.

- Входы SoftMax могут быть отрицательными. Просто вероятности этих классов будут ниже вероятностей остальных.
