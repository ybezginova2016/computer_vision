# Colored Images in Python

В трёхмерном массиве, полученном из изображения, всё так же: первая координата — это номер строки, вторая — номер столбца. Ещё добавляется третья координата — номер канала.



```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating 2-dimensional array
np.array([[0, 255],
                    [255, 0]])
```

```python
# creating 3-dimensional array
np.array([[[0, 255, 0], [128, 0, 255]], 
                    [[12, 89, 0], [5,  89, 245]]])
```

### Conclusion: Вместо изображения получается массив:

## Task 2


Изобразите на экране канал только с красным цветом.

```python
# extract a RED channel only
red_channel = array[:,:,0]
plt.imshow(red_channel)
plt.colorbar();
```

```python
# extract a GREEN channel only
green_channel = array[:,:,1]
plt.imshow(green_channel)
plt.colorbar();
```

```python
# extract a BLUE channel only
blue_channel = array[:,:,2]
plt.imshow(blue_channel)
plt.colorbar();
```
