где e — число Эйлера; приблизительно равно 2.718281828.

Это число в диапазоне от 0 до 1 можно трактовать как предсказание нейронной сети, к какому классу относится объект — отрицательному или положительному.

### Если вероятность правильного ответа p приблизительно равна единице, то -log(p) — положительное число, близкое к нулю. То есть ошибка маленькая. 

### Когда вероятность правильного ответа p≈0, то -log(p) — большое положительное число. Ошибка тоже большая.

- Сигмоидная функция в логистической регрессии нужна, чтобы привести ответы модели к диапазону от 0 до 1. Она нужна, чтобы получить вероятность положительного класса.
- Логистическая и линейная регрессии отличаются только функцией сигмоиды, которая не содержит параметры.
- Чем хуже модель, тем выше кросс-энтропия. ВСЕ — это функция потерь, её минимизируют, чтобы улучшить модель.
